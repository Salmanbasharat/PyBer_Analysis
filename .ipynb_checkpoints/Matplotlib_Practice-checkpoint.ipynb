```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add the label for the legend
plt.plot(x_axis, y_axis, label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=3, label='Boston')
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.xlim(0, 5)
plt.title("PyBer Fare by Month")
plt.grid()
plt.legend()
plt.figure(figsize=(20,10))
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["jan", "Feb", "Mar", "Apr", "May", "June", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(x_axis, y_axis, c='red', label="Chicago", marker='x')
plt.xlabel("Days")
plt.ylabel("Fare ($)")
plt.title("PyBer fares by Months")
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make Y axis data bigger
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i *5 for i in y_axis], c='skyblue')
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values=[0,0,0.1,0,0,0,0.3,0,0,0,0,0]
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%1.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
